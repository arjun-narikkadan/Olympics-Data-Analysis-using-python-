# OLYMPICS DATASET ANALYSIS

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```python
#load the dataset

athletes = pd.read_csv('E:/OLYMPIC DATASET/athlete_events.csv')
region = pd.read_csv('E:/OLYMPIC DATASET/noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
#join the dataframes 

athletes_df = athletes.merge(region, how = 'left', on = 'NOC')

athletes_df.head()
```

```python
athletes_df.shape
```

```python
#column name consistent

athletes_df.rename(columns = {'region': 'Region','notes':'Notes'},inplace = True)
```

```python
#when merged the column names are improper so used the rename function to make them in a proper way
```

```python
#top countries participating 

top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)

top_10_countries
```

```python
#barplot of the top 10 countries that participatedin the olympics 

plt.figure(figsize = (12,6))

#plt.xticks(rotation = 20)

plt.title('overall participation by country' )
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette ='Set2');
```

```python
#Age distribution of participants 

# histogram that shows the age distribution of the participants 

plt.figure(figsize = (12,6))
plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participents')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'orange',edgecolor = 'white');
```

```python
#diffrent winter olympics sports

winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#summer olympic sports

summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
#Male and Female participants 

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
#pie chart for male and female athletes 

plt.figure(figsize=(12,6))
plt.title("Gender Distribution")
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True);
```

```python
#Total medals athletes have won 

athletes_df.Medal.value_counts()
```

```python
#Total number of female participants in each olympics 

female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style = "darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data = womenOlympics, palette='Spectral')
plt.title("WomenOlympics")
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of female athletes over Time')
```

```python
#athletes who won gold medals

goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals.head()
```

```python
goldMedals.tail()
```

```python
#take only values that are diffrent from NaN

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#gold beyond  age 60

goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
#plot for sporting event 

plt.figure(figsize = (15,8))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medal for athletes over age 60')
```

```python
#Gold Medals earned by each country 

goldMedals.Region.value_counts().reset_index(name="Medal").head(5)
```

```python
goldMedals.Region.value_counts().reset_index(name="Medal").tail(5)
```

```python
# Dsiplaying the top 6 countries who got the most number of Medals 


TotalGoldMedals = goldMedals.Region.value_counts().reset_index(name="Medal").head(6)
g = sns.catplot(x = "index" , y ='Medal', data = TotalGoldMedals, height =8, kind ='bar', palette = 'rocket')
g.despine(left=True)
g.set_xlabels("top 6 contries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per country")
```

```python
# RIO Olympics 

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

```python
# displaying the top 20 nation who secured the most number of gold medals in the Olympics 

sns.barplot(x = team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016');
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize = (12,10))
axis = sns.scatterplot(x = 'Height', y = 'Weight', data = not_null_medals, hue = 'Sex')
plt.title('Height vs Weight of Olympic medalists')
```
